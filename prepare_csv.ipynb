```python
import pandas as pd
import numpy as np
import os
```

```python
finetune_data = pd.DataFrame()

with open('prompts.txt') as file:
    lines = file.readlines()
    prompts = [line.rstrip() for line in lines]

finetune_data['prompts'] = pd.DataFrame(pd.Series(prompts))
finetune_data
```

```python
with open('completion.txt', encoding="utf8") as file:
    lines = file.readlines()
    raw = [line.rstrip() for line in lines]

completion = ''
completions = []

for line in raw:

    if line == '####':
        if len(completion) is not 0:
            completions.append(completion)

        completion = ''
    else:
        completion = completion + line



finetune_data['completions'] = pd.DataFrame(pd.Series(completions))
finetune_data
```
